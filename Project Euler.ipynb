```r
##Question 3: The prime factors of 13195 are 5, 7, 13 and 29.

#What is the largest prime factor of the number 600851475143 ?


# Sieve of Eratosthenes for generating primes 2:n
esieve <- function(n) {
    if (n==1) return(NULL)
    if (n==2) return(n)
    # Create a list of consecutive integers {2,3,â€¦,N}.
    l <- 2:n
    # Start counter
    i <- 1
    # Select p as the first prime number in the list, p=2.
    p <- 2
    while (p^2<=n) {
        # Remove all multiples of p from the l.
        l <- l[l==p | l%%p!=0]
        # set p equal to the next integer in l which has not been removed.
        i <- i+1 # Repeat steps 3 and 4 until p2 > n, all the remaining numbers in the list are primes
        p <- l[i]
    }
    return(l)
    }
    
primefactor <- function (n) {
    factors <- c()
    primes <- esieve(floor(sqrt(n)))
    d <- which(n %% primes == 0)
    if (length(d) == 0)
        return(n)
    for (q in primes[d]) {
        while (n %% q == 0) {
            factors <- c(factors, q)
            n <- n/q } } 
    if (n > 1) factors <- c(factors, n)
    return(factors)
}
max(primefactor(600851475143))
```
